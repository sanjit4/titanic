```python
import numpy as np
import pandas as pd
```

```python
train_data = pd.read_csv("dataset/titanic/train.csv")
train_data.head()
```

```python
test_data = pd.read_csv("dataset/titanic/test.csv")
test_data.head()
```

```python
# data of women
women = train_data.loc[train_data.Sex == 'female']

# survived column of women
women_survived = women['Survived']

# rate of survival within women
# adding the ones in the column = total survivors
# total number of rows =  total women
women_survival_rate = sum(women_survived)/len(women_survived)
```

```python
men_survived = train_data.loc[train_data.Sex=='male']['Survived']
men_survival_rate = sum(men_survived)/len(men_survived)
```

```python
print("women survival rate:", women_survival_rate)
print("men survival rate:", men_survival_rate)
```

titles

```python
def substrings_in_string(big_string, substrings):
    # nan value is float type
    if isinstance(big_string, float):
        return np.nan
    for substring in substrings:
        if big_string.find(substring) != -1:
            return substring
    print(big_string)
    return np.nan
```

```python
title_list=['Mrs', 'Mr', 'Master', 'Miss', 'Major', 'Rev', 'Dr', 'Ms', 'Mlle', 'Col', 'Capt', 'Mme', 'Countess', 'Don', 'Jonkheer']
train_data['Title'] = train_data['Name'].map(lambda x: substrings_in_string(x, title_list))

def replace_titles(x):
    title = x['Title']
    if title in ['Don', 'Major', 'Capt', 'Jonkheer', 'Rev', 'Col']:
        return 'Mr'
    elif title in ['Countess', 'Mme']:
        return 'Mrs'
    elif title in ['Mlle', 'Ms']:
        return 'Miss'
    else:
        return title
    
train_data['Title'] = train_data.apply(replace_titles, axis=1)
```

```python
test_data['Title'] = test_data['Name'].map(lambda x: substrings_in_string(x, title_list))
test_data['Title'] = test_data.apply(replace_titles, axis=1)
```

deck

```python
cabin_list = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'T', 'Unknown']

train_data['Deck'] = train_data['Cabin'].map(lambda x: substrings_in_string(x, cabin_list))
print(train_data['Deck'].value_counts())
```

```python
test_data['Deck'] = test_data['Cabin'].map(lambda x: substrings_in_string(x, cabin_list))
print(test_data['Deck'].value_counts())
```

family size

```python
train_data['Family_Size'] = train_data['SibSp'] + train_data['Parch']
```

```python
test_data['Family_Size'] = test_data['SibSp'] + test_data['Parch']
```

Age*Class

```python
train_data['Age*Class']=train_data['Age']*train_data['Pclass']
```

```python
test_data['Age*Class']=test_data['Age']*test_data['Pclass']
```

fare per person

```python
train_data['Fare_Per_Person'] = train_data['Fare']/(train_data['SibSp']+train_data['Parch']+1)
```

```python
test_data['Fare_Per_Person'] = test_data['Fare']/(test_data['SibSp']+test_data['Parch']+1)
```

```python
from sklearn.ensemble import RandomForestClassifier

y_train = train_data['Survived']

features = ['Pclass', 'Sex', 'SibSp', 'Parch', 'Title', 'Deck', 'Family_Size', 'Age*Class', 'Fare_Per_Person']
X_train = pd.get_dummies(train_data[features])
X_test = pd.get_dummies(test_data[features])

model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': y_pred})
output.to_csv('submission.csv', index=False)
```
